Neste laboratório, veremos como abrir uma imagem, ver a largura, altura e quantidade de canais, bem como exibir a imagem.



---
1.   Abrindo uma imagem
2.   Exibindo a largura, altura e canais de uma imagem


Importando as bibliotecas necessárias ao projeto

```
from matplotlib import pyplot as plt
import cv2
```

Abrindo uma imagem com o OpenCV

```
image = cv2.imread("./Exemplo.png")
```

Exibindo a altura, largura e canais

```
print(image.shape)
#Outra forma é capturar os dados em variáveis separadas
(h, w, c) = image.shape
print("Largura: {} pixels".format(w))
print("Altura: {}  pixels".format(h))
print("Canais: {}".format(c))
```

Exibindo a imagem na tela e salvando a imagem como jpg

```
plt.imshow(image)
plt.title("Imagem exemplo")
plt.grid(False)
plt.show()
# O padrão de cores em geral é RGB, porém o OpenCV trata a imagem com os canais 
#de forma invertida, para melhorar a visualização, podemos colocar na ordem adequada
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title("Imagem exemplo em RGB")
plt.grid(False)
plt.show()
#Salvando a imagem
cv2.imwrite("exemplo1.jpg", image)
```
